```python
import torch
from torch.utils.data import Dataset, DataLoader

from tft_model import TFT
from tft_dataset import TimeSeriesDataset

# Define hyperparameters
batch_size = 32
learning_rate = 0.001
num_epochs = 50

# Load dataset
train_data = TimeSeriesDataset('train_data.csv')
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)

# Instantiate TFT model
tft_model = TFT(input_size=train_data.input_size, output_size=train_data.output_size)

# Define loss function and optimizer
loss_fn = torch.nn.MSELoss()
optimizer = torch.optim.Adam(tft_model.parameters(), lr=learning_rate)

# Train model
for epoch in range(num_epochs):
    for i, batch in enumerate(train_loader):
        inputs, targets = batch
        optimizer.zero_grad()
        outputs = tft_model(inputs)
        loss = loss_fn(outputs, targets)
        loss.backward()
        optimizer.step()
        print('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, i+1, len(train_loader), loss.item()))

# Save trained model
torch.save(tft_model.state_dict(), 'tft_model.pt')
```
